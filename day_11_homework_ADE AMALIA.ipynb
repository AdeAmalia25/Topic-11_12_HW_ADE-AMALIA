## No 3
#### Apartemen di Kota/Kabupaten mana yang memiliki rata-rata harga paling tinggi? Buatlah visualisasi untuk mendukung klaim Anda.

```python
df.groupby('Region')['CleanedPrice'].mean().sort_values(ascending=False)
```

```python
df_lookup = df.groupby(['Region'])[['CleanedPrice']].mean()
```

```python
df_lookup
```

```python
df_merged = df.merge(df_lookup, on=['Region'], suffixes=('_ori', '_mean'))
```

```python
df_merged.head(5)
```

```python
# code here
df.groupby('Region')['CleanedPrice'].mean().sort_values(ascending=False).plot(kind='bar', figsize=(10,5))
plt.title('Rata-rata harga apartemen berdasarkan region')
plt.xlabel('Region')
plt.ylabel('Rata-rata harga')
plt.xticks(rotation = 45) 
plt.show()

# terlihat Jakarta Timur memiliki rata-rata harga paling tinggi
```
